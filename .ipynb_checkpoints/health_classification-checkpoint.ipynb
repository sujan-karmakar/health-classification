```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, GradientBoostingClassifier, AdaBoostClassifier, StackingClassifier, VotingClassifier
```

```python
df = pd.read_csv("novagen_dataset.csv")
```

```python
df.isna().sum()
```

```python
df.sample(5)
```

```python
# Encoding
cols = ["Diet_Type__Vegan", "Diet_Type__Vegetarian", "Blood_Group_AB", "Blood_Group_B", "Blood_Group_O"]
num_cols = ["Age", "BMI", "Blood_Pressure", "Cholesterol", "Glucose_Level", "Heart_Rate", "Sleep_Hours", "Exercise_Hours", "Stress_Level"]

le = LabelEncoder()

for col in cols:
    df[col] = le.fit_transform(df[col])
    
df.head()
```

```python
X = df.drop("Target", axis = 1)
Y = df["Target"]

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, 
    test_size = 0.2,
    random_state = 42
)

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
lr = LogisticRegression(max_iter = 1000)
gnb = GaussianNB()
dtc = DecisionTreeClassifier()
knn = KNeighborsClassifier()
svc = SVC()
```

```python
# Heterogeneous Ensemble
stacking = StackingClassifier(
    estimators = [
        ("lr", lr),
        ("gnb", gnb),
        ("dtc", dtc),
        ("knn", knn),
        ("svc", svc)
    ]
)

voting = VotingClassifier(
    estimators = [
        ("lr", lr),
        ("gnb", gnb),
        ("dtc", dtc),
        ("knn", knn),
        ("svc", svc)
    ]
)
```

```python
# Bagging Classifiers
rfc = RandomForestClassifier(
    n_estimators = 601,
    max_depth = 6
)

base_model = LogisticRegression(max_iter = 1000)
bagging = BaggingClassifier(
    estimator = base_model,
    n_estimators = 201
)
```

```python
# Boosting Classifiers
gbc = GradientBoostingClassifier(
    n_estimators = 601,
    learning_rate = 0.01,
    max_depth = 6,
    subsample = 0.8
)

abc = AdaBoostClassifier(
    n_estimators = 601,
    learning_rate = 0.01
)
```

```python
models = {
    "Stacking Classifier": stacking,
    "Voting Classifier": voting,
    "Random Forest Classifier": rfc,
    "Bagging Classifier": bagging,
    "Gradient Boosting Classifier": gbc,
    "AdaBoost Classifier": abc
}
```

```python
model_performance = {
    "model": [],
    "precision": [],
    "recall": []
}

for model_name, model in models.items():
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    
    print(model_name, " : ")
    model_performance["model"].append(model_name)
    print(f"Accuracy Score : {accuracy_score(Y_test, Y_pred)}")
    print(f"Precision Score : {precision_score(Y_test, Y_pred)}")
    model_performance["precision"].append(float(f"{precision_score(Y_test, Y_pred):.2f}"))
    print(f"Recall Score : {recall_score(Y_test, Y_pred)}")
    model_performance["recall"].append(float(f"{recall_score(Y_test, Y_pred):.2f}"))
    print(f"F1 Score : {f1_score(Y_test, Y_pred)}")
    print(f"Classification Report : \n{classification_report(Y_test, Y_pred)}")
    print(f"Confusion Matrix : \n{confusion_matrix(Y_test, Y_pred)}")
    print("\n\n")
```

```python
evaluation = pd.DataFrame(model_performance)

evaluation
```

# Gradient Boosting Classifier performs best based on recall [First priority] and precision [Second priority]
